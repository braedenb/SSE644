#### Create a Drodown Widget containing every menu item.
key = workbook name

value = csv file name

```python
from collections import OrderedDict
from IPython.display import (display, clear_output)
from ipywidgets import Dropdown
# We create a DropdownWidget containing the keys (workbook name)
# and the values (csv file name) of every menu item.
dw = Dropdown(
    options = {'Top Countries':'world cup data.csv', 'Top Scorers':'world cup scorers.csv'},
    value = 'world cup data.csv',
    description = 'Excel workbook')
dw
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Read in the selected Excel workbook.

```python
wc = pd.read_csv(dw.value)
```

```python
if dw.value == 'world cup data.csv':
    countries = wc[0:5]['Country']
    points = wc[0:5]['Points']
    y_pos = np.arange(len(countries))
    plt.bar(y_pos, points, align='center', alpha=0.5)
    plt.xticks(y_pos, countries)
    plt.xlabel('Country')
    plt.ylabel('Points')
    plt.title('Top Points in World Cup')
else:
    players = wc[0:5]['Player']
    goals = wc[0:5]['Goals Scored']
    y_pos = np.arange(len(players))
    plt.bar(y_pos, points, align='center', alpha=0.5)
    plt.xticks(y_pos, players)
    plt.xlabel('Player')
    plt.ylabel('Goals')
    plt.title('Top Scorers in World Cup')
```

```python
if dw.value == 'world cup data.csv':
    wc = wc.sort_values('Points Per Match', ascending=False)
    countries = wc[0:8]['Country']
    ppm = wc[0:8]['Points Per Match']
    y_pos = np.arange(len(countries))
    plt.bar(y_pos, ppm, align='center', alpha=0.5)
    plt.xticks(y_pos, countries)
    plt.xlabel('Country')
    plt.ylabel('Points Per Match')
    plt.title('Top Points Per Match in World Cup')
else:
    wc = wc.sort_values('Goals Per Match', ascending=False)
    players = wc[0:8]['Player']
    gpm = wc[0:8]['Goals Per Match']
    y_pos = np.arange(len(players))
    plt.bar(y_pos, gpm, align='center', alpha=0.5)
    plt.xticks(y_pos, players)
    plt.xlabel('Player')
    plt.ylabel('Goals Per Match')
    plt.title('Top Goals Per Match in World Cup')
```

#### Time the sorting algorithm.

```python
if dw.value == 'world cup data.csv':
    %timeit wc.sort_values('Points Per Match', ascending=False)
else:
    %timeit wc.sort_values('Goals Per Match', ascending=False)
```
